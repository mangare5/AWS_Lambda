```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

%matplotlib inline
# fijo la semilla aleatoria por reproducibilidad
np.random.seed(7)
```

```python
pd.set_option('display.max_colwidth', None)

# Leemos los datos
datapath = 'results_for_comparison_GPTtest_multi_shot_EAMT.tsv'

df = pd.read_csv(datapath, index_col=None, encoding = "utf-8", sep='\t')
df.head()
```

```python
def count_words(string):
    # Split the string into words based on whitespace
    words = string.split()
    # Count the number of words
    num_words = len(words)
    return num_words
```

```python
lengths = []
for text in df.source:
  source_seg = text
  word_count = count_words(text)
  print("Number of words:", word_count)
  lengths.append(word_count)

df['source_length'] = lengths

print(df)
```

```python
# Define the function to calculate mean string length
def mean_string_length(strings):
    total_length = 0
    num_strings = len(strings)

    # Calculate the total length of all strings
    for string in strings:
        total_length += len(string)

    # Calculate the mean length
    if num_strings > 0:
        mean_length = total_length / num_strings
    else:
        mean_length = 0

    return mean_length

# Apply the function to the DataFrame columns
source_mean_length = mean_string_length(df['source'])

print("Source Mean length of strings in the DataFrame column:", source_mean_length)

target_mean_length = mean_string_length(df['trans'])

print("Target Mean length of strings in the DataFrame column:", target_mean_length)
```

```python
# Calcular la media de la DA obtenida
media = df['human_eval_old'].mean()

# Calcular la desviación estándar usando pandas
desviacion_estandar = df['human_eval_old'].std()

print("Media de los scores:", media)
print("Desviación estándar de los scores:", desviacion_estandar)
```
